# K_means utilizando Sklearn e dados randomicos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
np.random.seed(1)
a = np.random.normal(loc=10,scale=1,size=10),np.random.normal(loc=5,scale=1,size=10)
b = np.random.normal(loc=3,scale=1,size=10),np.random.normal(loc=2,scale=1,size=10)
c = np.concatenate((a,b),axis=1)
```

```python
plot_(c[0],c[1])
```

```python
kmeans = KMeans(n_clusters=2, n_init=10).fit(c.T)
kmeans.labels_
```

```python
plot_(c[0],c[1],kmeans.labels_)
```
